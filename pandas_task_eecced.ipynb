## 1. Обработать пропущенные значения, используя одну из следующих стратегий:
a Удалить строки, содержащие пропуски;

b Заменить минимальным значением;

c Заменить наиболее часто встречающимся значением (модой);

d Заменить медианой;

**e Заменить средним значением.**

Добавить в таблицу столбцы-индикаторы, сигнализирующие, где была произведена замена пропущенного значения. Имя столбца-индикатора сделать путем прибавления к имени заменяемого столбца суффикса `_indicator`. Сохранить полученные DataFrame-ы в csv-файлы без индексов, но с заголовками столбцов с префиксом `_stage1`, то есть people_stage1.csv, cars_stage1.csv, appartments_stage1.csv.

```python
people_indicators = people.isna().astype('int')
cars_indicators = cars.isna().astype('int')
apartments_indicators = apartments.isna().astype('int')
people_indicators = people_indicators.loc[:, people_indicators.any()]
cars_indicators = cars_indicators.loc[:, cars_indicators.any()]
apartments_indicators = apartments_indicators.loc[:, apartments_indicators.any()]

people = people.fillna(people.mean()).join(people_indicators, rsuffix='_indicator')
cars = cars.fillna(cars.mean()).join(cars_indicators, rsuffix='_indicator')
apartments = apartments.fillna(apartments.mean()).join(apartments_indicators, rsuffix='_indicator')

people.to_csv('data/pandas-task/eecced/people_stage1.csv', index=False)
cars.to_csv('data/pandas-task/eecced/cars_stage1.csv', index=False)
apartments.to_csv('data/pandas-task/eecced/appartments_stage1.csv', index=False)
```

## 2. В таблицу people добавить один из следующих столбцов:

## 3. Отсортировать таблицу cars, полученную в результате предыдущего пункта, следующим образом:
a По весу владельца, деленному на рост владельца по возрастанию;

b По возрасту владельца по возрастанию;

**c По стоимости автомобиля, деленной на мощность по убыванию;**

d По доходу владельца по убыванию;

e По стоимости по убыванию и после по мощности по возрастанию.

Сохранить данные с индексами и без заголовков в cars_stage3.csv

```python
price_to_power_ratio = (cars.Price / cars.Power).sort_values(ascending=False)
cars = cars.iloc[price_to_power_ratio.index]

cars.to_csv('data/pandas-task/eecced/cars_stage3.csv', header=False)
```

## 5. Выполнить следующее задание:
a Сохранить таблицу максимальных зарплат в зависимости от пола, профессии и марки машины.

b Сохранить таблицу средних зарплат в зависимости от пола, профессии и количества квартир.

c Сохранить таблицу минимальных зарплат в зависимости от пола, профессии и количества машин.

**d Сохранить таблицу минимальных и максимальных зарплат в зависимости от пола, профессии и количества машин.**

e Сохранить таблицу средних, минимальных и максимальных зарплат в зависимости от пола, профессии и количества квартир.

```python
people['Cars_count'] = 0
cars_count = cars.groupby('ID_person').size()
people.loc[cars_count.index, 'Cars_count'] = cars_count

salary_extremes = people.groupby(['Sex', 'Profession', 'Cars_count']).Salary.agg([min, max])

salary_extremes.to_csv('data/pandas-task/eecced/salary_extremes_stage5.csv')
```
