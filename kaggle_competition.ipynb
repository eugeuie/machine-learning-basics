```python
import pandas as pd


X = pd.read_csv('data/kaggle-competition/TrainX.csv', index_col='Id')
y = pd.read_csv('data/kaggle-competition/TrainY.csv', index_col='Id').Value
testX = pd.read_csv('data/kaggle-competition/TestX.csv', index_col='Id')

X = X[X <= 1]
X = X.fillna(value=X.mean())
```

```python
from sklearn.linear_model import PassiveAggressiveRegressor
from sklearn.kernel_ridge import KernelRidge
from sklearn.svm import SVR
from sklearn.ensemble import StackingRegressor


model = StackingRegressor([
    ('SVR', SVR()),
    ('KR', KernelRidge(kernel='rbf')),
    ('PAR', PassiveAggressiveRegressor(random_state=0))
    ])

model.fit(X, y)
y_predicted = model.predict(testX)
y_predicted
```

```python
y_predicted = pd.DataFrame({'Value' : y_predicted})
y_predicted['Id'] = range(len(y_predicted))
y_predicted.to_csv('data/kaggle-competition/Evgenia_Abdullaeva.csv', index=False)
```
